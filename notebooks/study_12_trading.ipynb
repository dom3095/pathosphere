# Study 12 — Trading Validation

```python
import os, sqlite3
from pathlib import Path

notebook_dir = Path.cwd()
while not (notebook_dir / 'data/db/pathosphere.db').exists():
    notebook_dir = notebook_dir.parent
    if notebook_dir == notebook_dir.parent: break
os.chdir(notebook_dir)

DB = Path('data/db/pathosphere.db').resolve()
CONN = sqlite3.connect(str(DB))
CONN.row_factory = sqlite3.Row
```

```python
trades = CONN.execute('SELECT COUNT(*) as c FROM trades').fetchone()['c']
ports = CONN.execute('SELECT COUNT(*) as c FROM portfolios').fetchone()['c']

print(f'Trades: {trades}')
print(f'Portfolios: {ports}')

if trades == 0:
    print('\n⬜ Trading not populated. Run: uv run pathos cycle')
else:
    print('\n✓ Trading data present')
```
