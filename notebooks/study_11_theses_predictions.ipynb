# Study 11 — Theses & Predictions Status

```python
import os, sqlite3
from pathlib import Path

notebook_dir = Path.cwd()
while not (notebook_dir / 'data/db/pathosphere.db').exists():
    notebook_dir = notebook_dir.parent
    if notebook_dir == notebook_dir.parent: break
os.chdir(notebook_dir)

DB = Path('data/db/pathosphere.db').resolve()
CONN = sqlite3.connect(str(DB))
CONN.row_factory = sqlite3.Row
print(f'CWD: {Path.cwd()}')
```

```python
# Theses + predictions
theses_total = CONN.execute('SELECT COUNT(*) as c FROM theses').fetchone()['c']
predictions_total = CONN.execute('SELECT COUNT(*) as c FROM predictions').fetchone()['c']
trades_total = CONN.execute('SELECT COUNT(*) as c FROM trades').fetchone()['c']
ports_total = CONN.execute('SELECT COUNT(*) as c FROM portfolios').fetchone()['c']

print(f'Theses: {theses_total}')
print(f'Predictions: {predictions_total}')
print(f'Trades: {trades_total}')
print(f'Portfolios: {ports_total}')

if theses_total + predictions_total + trades_total == 0:
    print('\n⬜ Phase 3 not yet populated. Run: uv run pathos cycle')
else:
    print('\n✓ Phase 3 data present')
```
